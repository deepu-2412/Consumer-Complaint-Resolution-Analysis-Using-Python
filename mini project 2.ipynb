```python
# Convert all negative days held to zero
train_data['Days held'] = train_data['Days held'].clip(lower=0)
```

```python
# Drop unnecessary columns for model building
train_data.drop(['Company', 'State', 'Year', 'Days held'], axis=1, inplace=True)
```

```python
# Change 'Consumer Disputed' column to 0 and 1
train_data['Consumer disputed?'] = train_data['Consumer disputed?'].map({'No': 0, 'Yes': 1})
```

```python
# Create dummy variables for categorical features
categorical_cols = ['Product', 'Submitted via', 'Company response to consumer', 'Timely response?']
train_data = pd.get_dummies(train_data, columns=categorical_cols, drop_first=True)
```

```python
# Separate the target variable from the features
X = train_data.drop('Consumer disputed?', axis=1)
y = train_data['Consumer disputed?']

y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()

X['Issue']=le.fit_transform(X['Issue'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Scaling the data sets
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
# Perform feature selection with PCA
pca = PCA(n_components=0.8)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
# Splitting the data sets into X and Y
X_train, X_val, y_train, y_val = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
lr=LogisticRegression()

dtc=DecisionTreeClassifier()

rf = RandomForestClassifier()

abc= AdaBoostClassifier()

gbc = GradientBoostingClassifier()

kc = KNeighborsClassifier()

xgb = XGBClassifier()
```

```python
# Define the models
models = [lr,dtc,rf,abc,gbc,kc,xgb]
```

```python
X_train
```

```python
model_names=[]
test_accuracies=[]
val_accuracies=[]


for model in models:
    model.fit(X_train, y_train)
    y_pred_test = model.predict(X_val)
    y_pred_train = model.predict(X_train)
    test_accuracy = accuracy_score(y_val, y_pred_test)
    train_accuracy = accuracy_score(y_train, y_pred_train)
    model_names.append(model.__class__.__name__)
    test_accuracies.append(test_accuracy)
    val_accuracies.append(train_accuracy)
```

```python
# Compare model accuracies
accuracy_df = pd.DataFrame({'Model': model_names, 'Test Accuracy': test_accuracies, 'Train Accuracy': val_accuracies})
print(accuracy_df)
```

```python
# Choose the model with the highest test accuracy
best_model_index = accuracy_df['Test Accuracy'].idxmax()
best_model = models[best_model_index]
```

```python
# Fit the best model on the entire dataset
best_model.fit(X_pca, y)
```
